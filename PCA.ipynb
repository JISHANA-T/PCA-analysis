```python


import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
X = z.loc[:, 'x1':'x28']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA(n_components=28)
z_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(1, pca.n_components_+1), pca.explained_variance_ratio_, marker='o', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('PCA Scree Plot')
plt.grid()
plt.show()
```

```python
# random data
data = np.random.rand(100, 2)
from sklearn.cluster import KMeans


# K-Means with 3 clusters
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)

# cluster centroids and labels
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

# Plot the data and centroids
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='rainbow')
plt.scatter(centroids[:, 0], centroids[:, 1], color='black', marker='X', s=100)
plt.show()
```
